```python
"""
===========================================================
📘 Data Scaling, Normalization & Transformation - Full Guide
===========================================================

🔹 INTRODUCTION
Data preprocessing is a critical step in machine learning.

👉 Definition:
Data Scaling, Normalization, and Transformation are techniques used to
prepare data by ensuring features are on similar scale and distribution.

Why?
Many ML algorithms (KNN, SVM, Neural Networks) are sensitive to feature scale.

Problem:
Income → 10,000 to 1,000,000
Age → 0 to 100

👉 Income dominates → Model becomes biased ❌

-----------------------------------------------------------
🔹 TYPES OF TECHNIQUES
-----------------------------------------------------------
1. Scaling (Feature-based)
2. Normalization (Row-based)
3. Transformation (Distribution change)
"""

# =========================================================
# 🔹 1. MIN-MAX SCALING
# =========================================================

"""
Definition:
Rescales data to a fixed range [0,1]

Formula:
X_scaled = (X - X_min) / (X_max - X_min)

Use Case:
✔ When range is known
❌ Not good for outliers
"""

from sklearn.preprocessing import MinMaxScaler
import pandas as pd

df = pd.DataFrame({
    'Age': [25, 30, 35, 40],
    'Income': [50000, 60000, 70000, 80000]
})

minmax = MinMaxScaler()
df_minmax = pd.DataFrame(minmax.fit_transform(df), columns=df.columns)

print("\nMin-Max Scaling:\n", df_minmax)


# =========================================================
# 🔹 2. STANDARDIZATION (Z-SCORE)
# =========================================================

"""
Definition:
Transforms data → Mean = 0, Std = 1

Formula:
Z = (X - Mean) / Std

Use Case:
✔ Most ML models (SVM, Logistic Regression)
✔ Works well with normal distribution
"""

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standard = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

print("\nStandardization:\n", df_standard)


# =========================================================
# 🔹 3. ROBUST SCALING
# =========================================================

"""
Definition:
Uses Median and IQR → robust to outliers

Formula:
(X - Median) / IQR

Use Case:
✔ Data with outliers
"""

from sklearn.preprocessing import RobustScaler

df_outlier = pd.DataFrame({
    'Feature1': [10, 20, 30, 1000],
    'Feature2': [5, 7, 9, -999]
})

robust = RobustScaler()
df_robust = pd.DataFrame(robust.fit_transform(df_outlier), columns=df_outlier.columns)

print("\nRobust Scaling:\n", df_robust)


# =========================================================
# 🔹 4. LOG TRANSFORMATION
# =========================================================

"""
Definition:
Applies log function to reduce skewness

Use Case:
✔ Right-skewed data (income, population)

Note:
✔ Only works for positive values
"""

import numpy as np

df_log = pd.DataFrame({
    'Income': [10000, 50000, 100000, 1000000]
})

df_log['Log_Income'] = np.log(df_log['Income'])

print("\nLog Transformation:\n", df_log)


# =========================================================
# 🔹 5. POWER TRANSFORMATIONS
# =========================================================

"""
Definition:
Transforms data to make it more normal-like

Types:
✔ Box-Cox → Only positive data
"""

from sklearn.preprocessing import PowerTransformer

# Box-Cox
data_pos = np.random.rand(100, 1) + 1
boxcox = PowerTransformer(method='box-cox')
print("\nBox-Cox:\n", boxcox.fit_transform(data_pos)[:5])


# =========================================================
# 🔹 6. NORMALIZATION
# =========================================================

"""
Definition:
Scales each row so that its norm = 1

Difference:
Scaling → column-wise
Normalization → row-wise
"""

from sklearn.preprocessing import Normalizer

X = np.array([[10, 20, 30]])

# L1 Normalization
"""
Definition:
Sum of absolute values = 1
"""
l1 = Normalizer(norm='l1')
print("\nL1 Normalization:\n", l1.transform(X))

# L2 Normalization
"""
Definition:
Vector length = 1
"""
l2 = Normalizer(norm='l2')
print("\nL2 Normalization:\n", l2.transform(X))


# =========================================================
# 🔹 FINAL SUMMARY
# =========================================================

"""
Technique        Use Case
---------------------------------------
Min-Max          Fixed range
Standardization  Normal distribution
Robust Scaling   Outliers
Log Transform    Skewed data
Box-Cox          Positive skewed data
L1 Norm          Sparse data
L2 Norm          Distance-based models

-----------------------------------------------------------
🎯 CONCLUSION

✔ Scaling improves accuracy
✔ Helps faster convergence
✔ Makes models fair

👉 "Good preprocessing = Better model performance"
"""
```
